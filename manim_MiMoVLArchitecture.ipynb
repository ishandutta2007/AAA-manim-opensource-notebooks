```python
from manim import *
from manim import Scene, Text, Rectangle
from manim.mobject.geometry.line import Arrow
```

```python
%%manim -qm -v WARNING MiMoVLArchitecture

from manim import Scene, Text, Rectangle
from manim.mobject.geometry.line import Arrow

class MiMoVLArchitecture(Scene):
    def construct(self):
        # Title
        title = Text("MiMo-VL-7B Architecture", font_size=36).to_edge(UP)
        self.add(title)

        # Input image
        input_image = Text("Input Image", color=WHITE).to_edge(LEFT)
        self.play(Write(input_image))

        # Vision Encoder (ViT)
        vit = Rectangle(width=2, height=1, color=BLUE).next_to(input_image, RIGHT)
        vit_text = Text("Vision Encoder (ViT)", color=BLUE).next_to(vit, DOWN)
        vit_details = Text("Qwen2.5-ViT, 32 layers, 16 heads, 1280 hidden size", color=BLUE).next_to(vit_text, DOWN)
        self.play(Create(vit), Write(vit_text), Write(vit_details))
        # Use Create to animate the drawing of the arrow
        self.play(Create(Arrow(start=input_image.get_center(), end=vit.get_center(), color=WHITE)))

        # Projector (MLP)
        mlp = Rectangle(width=2, height=1, color=GREEN).next_to(vit, RIGHT)
        mlp_text = Text("Projector (MLP)", color=GREEN).next_to(mlp, DOWN)
        mlp_details = Text("Randomly initialized, warmed up in Stage 1", color=GREEN).next_to(mlp_text, DOWN)
        self.play(Create(mlp), Write(mlp_text), Write(mlp_details))
        # Use Create to animate the drawing of the arrow
        self.play(Create(Arrow(start=vit.get_center(), end=mlp.get_center(), color=WHITE)))

        # Language Model (LLM)
        llm = Rectangle(width=2, height=1, color=RED).next_to(mlp, RIGHT)
        llm_text = Text("Language Model (LLM)", color=RED).next_to(llm, DOWN)
        llm_details = Text("MiMo-7B, 36 layers, 32 heads, 4096 hidden size", color=RED).next_to(llm_text, DOWN)
        self.play(Create(llm), Write(llm_text), Write(llm_details))
        # Use Create to animate the drawing of the arrow
        self.play(Create(Arrow(start=mlp.get_center(), end=llm.get_center(), color=WHITE)))

        # Output text
        output_text = Text("Output Text", color=WHITE).to_edge(RIGHT)
        self.play(Write(output_text))
        # Use Create to animate the drawing of the arrow
        self.play(Create(Arrow(start=llm.get_center(), end=output_text.get_center(), color=WHITE)))

        # Wait for a bit
        self.wait(2)
```
