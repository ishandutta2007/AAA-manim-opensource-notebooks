```python
from manim import Scene, Text, Rectangle
from manim.mobject.geometry.line import Arrow
from manim import *
```

```python
%%manim -qm -v WARNING MiMoVLArchitecture

from manim import Scene, Text, Rectangle
from manim.mobject.geometry.line import Arrow
import numpy as np

class MiMoVLArchitecture(Scene):
    def construct(self):
        DEFAULT_REC_WIDTH=2
        # Title
        title = Text("MiMo-VL-7B Architecture", font_size=36).to_edge(UP)
        self.add(title)

        # Input image
        input_image = Text("Input", color=WHITE, width=DEFAULT_REC_WIDTH).next_to(title, 10*DOWN).to_edge(LEFT)
        self.play(Write(input_image))

        # Vision Encoder (ViT)
        vit = Rectangle(width=DEFAULT_REC_WIDTH, height=1, color=BLUE).next_to(input_image, 4*RIGHT)
        vit_text = Text("Vision Encoder (ViT)", color=BLUE, width=DEFAULT_REC_WIDTH).next_to(vit, DOWN)
        vit_details1 = Text("Qwen2.5-ViT,", color=BLUE, font_size=DEFAULT_FONT_SIZE//3).next_to(vit_text, DOWN)
        vit_details2 = Text("32 layers, 16 heads,", color=BLUE, font_size=DEFAULT_FONT_SIZE//3).next_to(vit_details1, DOWN)
        vit_details3 = Text("1280 hidden size", color=BLUE, font_size=DEFAULT_FONT_SIZE//3).next_to(vit_details2, DOWN)
        self.play(Create(vit), Write(vit_text), Write(vit_details1), Write(vit_details2), Write(vit_details3))
        self.play(Create(Arrow(start=input_image.get_right()-  np.array([0.1, 0, 0]), end=vit.get_left()+  np.array([0.3, 0, 0]), color=WHITE)))

        # Projector (MLP)
        mlp = Rectangle(width=DEFAULT_REC_WIDTH, height=1, color=GREEN).next_to(vit, 2*RIGHT)
        mlp_text = Text("Projector (MLP)", color=GREEN, width=DEFAULT_REC_WIDTH).next_to(mlp, DOWN)
        mlp_details1 = Text("Randomly initialized", color=GREEN, font_size=DEFAULT_FONT_SIZE//3).next_to(mlp_text, DOWN)
        mlp_details2 = Text("warmed up in Stage 1", color=GREEN, font_size=DEFAULT_FONT_SIZE//3).next_to(mlp_details1, DOWN)
        self.play(Create(mlp), Write(mlp_text), Write(mlp_details1), Write(mlp_details2))
        self.play(Create(Arrow(start=vit.get_right()-  np.array([0.3, 0, 0]), end=mlp.get_left()+  np.array([0.3, 0, 0]), color=WHITE)))

        # Language Model (LLM)
        llm = Rectangle(width=DEFAULT_REC_WIDTH, height=1, color=RED).next_to(mlp, 2*RIGHT)
        llm_text = Text("Language Model (LLM)", color=RED, width=DEFAULT_REC_WIDTH).next_to(llm, DOWN)
        llm_details1 = Text("MiMo-7B", color=RED, font_size=DEFAULT_FONT_SIZE//3).next_to(llm_text, DOWN)
        llm_details2 = Text("36 layers, 32 heads,", color=RED, font_size=DEFAULT_FONT_SIZE//3).next_to(llm_details1, DOWN)
        llm_details3 = Text("4096 hidden size", color=RED, font_size=DEFAULT_FONT_SIZE//3).next_to(llm_details2, DOWN)
        self.play(Create(llm), Write(llm_text), Write(llm_details1), Write(llm_details2), Write(llm_details3))
        self.play(Create(Arrow(start=mlp.get_right()-  np.array([0.3, 0, 0]), end=llm.get_left() +  np.array([0.3, 0, 0]), color=WHITE)))

        # Output text
        output_text = Text("Output", color=WHITE, width=DEFAULT_REC_WIDTH).next_to(llm, 2*RIGHT).to_edge(RIGHT)
        self.play(Write(output_text))
        self.play(Create(Arrow(start=llm.get_right(), end=output_text.get_left(), color=WHITE)))

        # Wait for a bit
        self.wait(2)
```
